```python
# fire weather dataframe
fwdf = pd.read_sql_query(''' 
    SELECT aggregate_data."index" as agg_idx, weather_data."index" as w_idx, aggregate_data.*, weather_data.*
    FROM aggregate_data
             JOIN weather_data
                  ON "CS-USAF" = stn AND julianday(date) between julianday(DISCOVERY_DATE) and julianday(CONT_DATE)
    WHERE FIRE_YEAR = 2014 and stn != 999999 ORDER BY aggregate_data.FIRE_NAME desc, weather_data.date; 
''', conn)
```
